```python
from pyspark.sql.functions import pandas_udf
from pyspark.sql.types import DoubleType
import pandas as pd
```

```python
data = [
    (1, 10.0, 5.0),
    (2, 20.0, 3.5),
    (3, 30.0, 7.2)
]

df = spark.createDataFrame(data, ["id", "value1", "value2"])
df.show()
```

```python
@pandas_udf(DoubleType())
def sum_values(v1: pd.Series, v2: pd.Series) -> pd.Series:
    return v1 + v2
```

```python
result_df = df.withColumn("sum", sum_values(df["value1"], df["value2"]))
result_df.show()
```

### Without defining udf, faster

```python
display(df.withColumn("sum", df.value1 + df.value2).collect())
```
