```python
from pyspark.sql.functions import col, count, count_if
from databricks_sample_package.udfs.functions import is_valid_email

df = (
    spark.read.table("sample_users_mar_19_1841")
    .withColumn("valid_email", is_valid_email(col("email")))
    .groupBy(col("user_type"))
    .agg(
        count("user_id").alias("total_count"),
        count_if("valid_email").alias("count_valid_emails")
    )
)

display(df)
```
